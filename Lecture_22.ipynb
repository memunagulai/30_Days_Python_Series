1️⃣ Local Variable

- Defined inside a function

- Accessible only within that function

- Example: temporary storage inside a calculation

```python
def greet():
    message = "Hello, World!"  # local variable
    print(message)

greet()
# print(message)  # Error: message is local
```

2️⃣ Global Variable

- Defined outside any function

- Accessible throughout the program

```python
count = 10  # global variable

def show_count():
    print("Count is:", count)

show_count()
```

3️⃣ global Keyword

- Allows modifying global variables inside a function



```python
counter = 0

def increment():
    global counter
    counter += 1 # counter = counter+1

increment()
increment()
print("Counter after increments:", counter)
```

```python
x = 10   # global variable

def change_value():
    global x
    x = 20   # modifying global variable

change_value()
print(x)
```

4️⃣ globals() Function

- Returns dictionary of all global variables

```python
x=10

def value():
    x=50 
    print("Local variable",x)
    a= globals()['x']
    print("inside function",a)

value()
print("Global variable",x)
```

```python
x = 100
y = 200

print(globals()['x'])  # shows all global variables including x, y
```

5️⃣ Recursion

- Function calls itself

- Must have base condition to stop recursion

- Useful for factorial, Fibonacci, tree traversal

```python
def my_func():  
    print("Memuna Gul")
    my_func()

my_func()
```

```python
i=0
def my_fun():
    global i 
    i=i+1
    print("My Function",i )
    my_fun()

my_fun()
```

```python
import sys 
print(sys.getrecursionlimit())

# sys.setrecursionlimit(1000)
```

6️⃣ Pass by Object Reference

- Python passes reference of the object, not a copy.

- If the object is mutable (like list), changes affect the original.

- If the object is immutable (like int, string), changes do not affect the original.

- Function and variable point to the same object in memory.

```python
#Mutable Object(list)

def add_element(lst):
    lst.append(100)
    print(lst, id(lst))

my_list = [1, 2, 3]
add_element(my_list)
print("Modified List:", my_list, id(my_list))
```

```python
# Immutable Object(integer)
def add_number(x):
    x = x + 5
    print(x, id(x))

num = 10
add_number(num)

print(num, id(num))
```

7️⃣ Function Decorator

- A function that modifies another function

- Used for logging, authorization, timing

```python
def decor(fun):
    def inner():
        print("IF: Before enhancing Function")
        fun()
        print("IF: After Enhancing Function")
    return inner

def num():
    print("We will use this function")
    print("and will enhance this in decorator")

result_fun = decor(num)
result_fun()
```

```python
def decor(fun):
    def inner():
        print("IF: Before enhancing Function")
        fun()
        print("IF: After Enhancing Function")
    return inner

@decor
def num():
    print("We will use this function")
    print("and will enhance this in decorator")

num()
```

```python
def decorator(func):
    def wrapper():
        print("Before function call")
        func()
        print("After function call")
    return wrapper

@decorator
def greet():
    print("Hello!")

greet()
```

```python
def decor(fun):
    def inner():
        a = fun()
        add = a + 5
        return add
    return inner

def num():
    return 10

result_fun = decor(num)
print(result_fun())
```

```python
def decor(num):
    def inner():
        a = num()
        add = a + 5
        return add
    return inner

@decor
def num():
    return 10

num()
```

8️⃣ Passing Array to Function

- Arrays can be passed like any object

```python
from array import * 
def print_array(arr):
    print(type(arr))
    for i in arr:
        print(i, end=" ")

nums = array('i',[1, 2, 3, 4])
print_array(nums)
```

9️⃣ Returning Array

- Function can return a list/array

```python
def create_array(n):
    return [i*2 for i in range(n)]

arr = create_array(5)
print("Returned Array:", arr)
```

🔟 Passing & Returning NumPy Array

- NumPy arrays are efficient for math operations

```python

import numpy as np

def multiply_array(arr):
    return arr * 2

a = np.array([1,2,3])
print("Original Array:", a)
print("Modified Array:", multiply_array(a))
```
