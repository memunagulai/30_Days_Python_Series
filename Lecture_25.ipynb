# Lecture 25 – Generators

1️⃣ What is a Generator

A generator is a special type of function that produces values one at a time instead of returning all values at once.

Generators use the `yield` keyword instead of `return`.

Real-life example:

Think of a water tap 🚰

Function → gives you all water at once (bucket)

Generator → gives water drop by drop (on demand)

2️⃣ Difference Between Function & Generator

| Function                  | Generator                  |
| ------------------------- | -------------------------- |
| Uses `return`             | Uses `yield`               |
| Executes fully            | Pauses and resumes         |
| Stores all data in memory | Produces values one by one |
| Returns a value           | Returns an iterator        |


```python
# Normal function
def square_list(nums):
    result = []
    for n in nums:
        result.append(n ** 2)
    return result

print(square_list([1, 2, 3, 4]))
```

```python
# Generator function
def square_generator(nums):
    for n in nums:
        yield n ** 2

gen = square_generator([1, 2, 3, 4])
print(gen)  # generator object
print(type(gen))
```

```python

# Generator function
def square_generator(nums):
    for n in nums:
        yield n ** 2

gen = square_generator([1, 2, 3, 4])
print(list(gen))  # generator object
print(type(gen))

```

3️⃣ `yield` Keyword

`yield`:

- Returns a value

- Pauses the function

- Remembers the state

- Continues from where it left off

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

```

4️⃣ Creating a Generator

```python
numbers = count_up_to(5)

print(next(numbers))
print(next(numbers))
print(next(numbers))
```

5️⃣ `next() `Function

`next()` is used to manually get the next value from a generator.

If no values are left, it raises `StopIteration`.

```python
print(next(numbers))
print(next(numbers))
```

6️⃣ Iteration Using `for` Loop

The safest and most common way to use generators.

```python
for num in count_up_to(5):
    print(num)
```

7️⃣ Generator Expressions

Short and clean way to create generators (like list comprehensions).


(expression for item in iterable)

```python
gen = (x ** 2 for x in range(1, 6))

for value in gen: 
    print(value)
```


8️⃣ Advantages of Generators

- Memory efficient

- Faster for large data

- Lazy evaluation

- Ideal for streams & big files

9️⃣ Memory Efficiency

```python
# List (stores all values)
numbers_list = [x for x in range(1, 1000000)]

# Generator (creates values on demand)
numbers_gen = (x for x in range(1, 1000000))

print(numbers_list[:5])
print(next(numbers_gen))
```

Real-life analogy:

List → whole book in memory 📚

Generator → reading page by page 📄

📝 Practice Questions

1. Create a generator that yields even numbers up to 20

2. Use `next()` to fetch first 3 values from a generator

3. Convert a list comprehension into a generator expression

4. Write a generator for Fibonacci series

5. Explain why generators are memory efficient

```python
# =========================
# Solution 1
# =========================
# Generator yielding even numbers up to 20
def even_numbers():
    for i in range(2, 21, 2):
        yield i

print("Solution 1:", list(even_numbers()))


# =========================
# Solution 2
# =========================
# Using next() to fetch first 3 values
gen = even_numbers()
print("Solution 2:")
print(next(gen))
print(next(gen))
print(next(gen))


# =========================
# Solution 3
# =========================
# List comprehension vs Generator expression
nums = [1, 2, 3, 4, 5]

list_comp = [x ** 2 for x in nums]
gen_exp = (x ** 2 for x in nums)

print("Solution 3:")
print("List:", list_comp)
print("Generator:", list(gen_exp))


# =========================
# Solution 4
# =========================
# Fibonacci generator
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

print("Solution 4:", list(fibonacci(10)))


# =========================
# Solution 5
# =========================
# Demonstrating memory efficiency concept
large_list = [x for x in range(1, 100000)]
large_generator = (x for x in range(1, 100000))

print("Solution 5:")
print("First 5 from list:", large_list[:5])
print("First 5 from generator:", [next(large_generator) for _ in range(5)])
```
