# Lecture 27 – Error/Exception Handling

1️⃣ Types of Errors

Errors in Python can be classified into three types:

🔹 Syntax Error

- Happens when Python cannot understand the code

- Usually due to typos or missing symbols


```python
print("Hello World"  # missing closing parenthesis
```

🔹 Runtime Error

- Occurs when the code runs but fails

- Examples: dividing by zero, accessing invalid index

```python
a = 10 / 0
```

🔹 Logical Error

- Code runs without crashing but gives wrong output

- Hardest to detect

```python
# Wrong formula for area of rectangle
length = 5
width = 10
area = length + width   # should be length * width
print(area)
```

2️⃣ try Block

- Wrap risky code inside `try`

- Python will catch exceptions instead of crashing

```python
try:
    num = int(input("Enter a number: "))
    print(10 / num)
except ZeroDivisionError:
    print("Cannot divide by zero!")
```

3️⃣ `except` Block

- Code inside `except` runs if an exception occurs

- You can specify exception type or leave it generic

```python
try:
    print(10 / 0)
except ZeroDivisionError:
    print("ZeroDivisionError caught!")
except Exception as e:
    print("Some other error:", e)
```

4️⃣ Multiple `except`

Handle different exceptions separately

```python
try:
    a = int("abc")
except ValueError:
    print("ValueError caught!")
except ZeroDivisionError:
    print("ZeroDivisionError caught!")
```

5️⃣ `else` Block

Runs only if no exception occurs

```python
try:
    result = 10 / 2
except ZeroDivisionError:
    print("Cannot divide by zero")
else:
    print("Division successful, result:", result)
```

6️⃣ `finally` Block

- Runs always, whether exception occurs or not

- Used for cleanup like closing files

```python
try:
    num = int(input("Enter a number: "))
    print(10 / num)
except ZeroDivisionError:
    print("Cannot divide by zero")
finally:
    print("This always runs")
```

7️⃣ `raise` Keyword

- Manually `raise` exceptions

- Useful to enforce rules or conditions

```python
age = -5
if age < 0:
    raise ValueError("Age cannot be negative")
```

8️⃣ Custom Exception

Define your own exception class

```python
class TooYoungError(Exception):
    pass

age = 12
try:
    if age < 18:
        raise TooYoungError("You are too young!")
except TooYoungError as e:
    print("Custom Exception:", e)
```

📝 Practice Questions

1. Write a program to handle division by zero using `try-except`

2. Handle multiple exceptions: ValueError and ZeroDivisionError

3. Write a program using `else` block to print success if no exception occurs

4. Use `finally` to close a file safely

5. Create a custom exception for negative salary and raise it

```python
# =========================
# Solution 1
# =========================
# Handle division by zero
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Solution 1: Cannot divide by zero!")


# =========================
# Solution 2
# =========================
# Handle multiple exceptions: ValueError and ZeroDivisionError
try:
    num = int("abc")    # ValueError
    result = 10 / num   # ZeroDivisionError
except ValueError:
    print("Solution 2: ValueError caught!")
except ZeroDivisionError:
    print("Solution 2: ZeroDivisionError caught!")


# =========================
# Solution 3
# =========================
# Using else block
try:
    result = 10 / 2
except ZeroDivisionError:
    print("Solution 3: Cannot divide by zero")
else:
    print("Solution 3: Division successful, result =", result)


# =========================
# Solution 4
# =========================
# Using finally block to simulate cleanup
try:
    num = int(input("Enter a number: "))
    print(10 / num)
except ZeroDivisionError:
    print("Solution 4: Cannot divide by zero")
finally:
    print("Solution 4: This always runs")


# =========================
# Solution 5
# =========================
# Using raise to manually trigger exception
age = -5
try:
    if age < 0:
        raise ValueError("Solution 5: Age cannot be negative")
except ValueError as e:
    print(e)


# =========================
# Solution 6
# =========================
# Custom exception
class TooYoungError(Exception):
    pass

age = 12
try:
    if age < 18:
        raise TooYoungError("Solution 6: You are too young!")
except TooYoungError as e:
    print(e)
```
