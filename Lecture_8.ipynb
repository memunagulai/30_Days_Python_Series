## 1) Relational / Comparison Operators
Operators that compare values and return a boolean (True / False).

- `==` : equal to
- `!=` : not equal to
- `>`  : greater than
- `<`  : less than
- `>=` : greater than or equal to
- `<=` : less than or equal to

**Real-life examples**
- Is your marks >= passing_marks?
- Is product price <= your budget?


```python
# Relational Operators examples
marks = 82
passing_marks = 50
print("Did the student pass?", marks >= passing_marks)  # True

price = 1200
budget = 1500
print("Can I buy it?", price <= budget)  # True

a= 10
b= 20
print("a == b ->", a == b)
print("a != b ->", a != b)
print("a > b ->", a > b)
print("a < b ->", a < b)
```

## 2) Logical Operators
Combine boolean conditions:

- `and` : True if both conditions are True
- `or`  : True if at least one condition is True
- `not` : negates the condition

**Real-life examples**
- Eligible for exam: (has_roll_number AND fees_paid)
- Can buy: (have_cash OR have_credit_card)


## Truth Tables for Logical Operators

Logical operators (`and`, `or`, `not`) are used to combine or modify conditions.

---

### ‚úî AND Operator (`and`)
The result is **True only if both conditions are True**.

| A     | B     | A and B |
|-------|--------|---------|
| True  | True   | True    |
| True  | False  | False   |
| False | True   | False   |
| False | False  | False   |

---

### ‚úî OR Operator (`or`)
The result is **True if at least one condition is True**.

| A     | B     | A or B |
|-------|--------|--------|
| True  | True   | True   |
| True  | False  | True   |
| False | True   | True   |
| False | False  | False  |

---

### ‚úî NOT Operator (`not`)
The result is the **opposite of the value**.

| A     | not A |
|--------|--------|
| True   | False |
| False  | True  |


```python
age = 20
has_id = True
print("Can apply for driving license?", age >= 18 and has_id)


has_cash = False
has_card = True
print("Can buy laptop?", has_cash or has_card)


is_raining = True
print("Go outside?", not is_raining)  # False

```

## 3) Assignment Operators
Store/modify variable values in-place:

- `=`   : simple assignment
- `+=`  : add then assign
- `-=`  : subtract then assign
- `*=`  : multiply then assign
- `/=`  : divide then assign
- `//=` : floor divide then assign
- `%=`  : modulus then assign
- `**=` : power then assign

**Real-life examples**
- Wallet balance updated after spending: `balance -= expense`
- Add monthly salary: `balance += salary`


```python
balance = 1000
print("Initial:", balance)

balance += 500   # salary
print("After salary:", balance)

balance -= 120   # groceries
print("After groceries:", balance)

score = 10
score *= 2
print("Doubled score:", score)
```
