# Lecture 23 – Modules and Packages

1️⃣ What is a Module

A module in Python is a file containing Python code (functions, classes, variables) that can be reused in other programs.
Modules help organize code, avoid repetition, and make programs easier to maintain.

Real-life example:
Think of a module like a toolkit. If you want to build furniture, you don’t make a new hammer every time you just use one from your toolkit.

2️⃣ Types of Modules

- Built-in Modules – Modules that come pre-installed with Python.

 Examples:  math, random, datetime, os

- User-Defined Modules – Modules created by the programmer for their own project.

3️⃣ `import` Statement

Used to bring a module into your program so you can use its functions, classes, or variables.

```python
# Using the math module
import math

print(math.sqrt(16))   # Output: 4.0
print(math.factorial(5))  # Output: 120
```

Real-life example:

If you need a map, you “import” your GPS app instead of drawing one yourself.

4️⃣ from … import Statement

Used to import specific functions or variables from a module.

```python
from math import pi, pow

print(pi)      # Output: 3.141592653589793
print(pow(2, 3))  # Output: 8
```

5️⃣ as Keyword

Used to give a module or function a nickname for easier usage.

```python
import numpy as np

arr = np.array([1, 2, 3])
print(arr)
```

Real-life example:

You can call “International Business Machines” just IBM.

6️⃣ Creating a User-Defined Module

Create a Python file, e.g., `my_module.py`

```python
# my_module.py

def greet(name):
    return f"Hello, {name}!"

pi_value = 3.14159
```

Import it in another file:

```python
# main.py

import my_package.my_module as my_module

print(my_module.greet("Memuna"))  # Output: Hello, Memuna!
print(my_module.pi_value)         # Output: 3.14159
```

7️⃣ What is a Package

A package is a collection of Python modules grouped in a folder.
It allows better organization of large projects.

Structure Example:

```python
my_package/
    __init__.py
    module1.py
    module2.py
```

8️⃣ Creating a Package

Create a folder, e.g., my_package

Add `__init__.py` (can be empty or used for initialization)

Add module files: module1.py, module2.py

```python
# module1.py
def add(a, b):
    return a + b

# module2.py
def multiply(a, b):
    return a * b
```

9️⃣ Importing from a Package

```python
from my_package.my_module import pi_value
# from my_package.module2 import multiply

print(pi_value)
# print(add(2, 3))        # Output: 5
# print(multiply(4, 5))   # Output: 20
```

1️⃣0️⃣ Advantages of Modules & Packages

- Code Reusability – Write once, use multiple times

- Better Organization – Split large programs into small files

- Maintainability – Easy to debug and update

- Namespace Management – Avoids naming conflicts

Real-life example:

Think of a library: each book is a module, and the library itself is a package.
