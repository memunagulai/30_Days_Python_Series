1️⃣ Variable Length Arguments (`*args`)

- Use `*args` to accept any number of positional arguments.

- Inside the function, `args` is a tuple.

Real-life analogy:

Think of it like a shopping cart — you can add any number of items.

```python
def add_numbers(*nums):
    total = 0
    for n in nums:
        total += n
    return total

print(add_numbers(10, 20, 30))  # 60
print(add_numbers(5, 15))       # 20
```

2️⃣ Keyword Variable Length Arguments (`**kwargs`)

- Use `**kwargs` to accept any number of keyword arguments.

- Inside the function, `kwargs` is a dictionary.

Analogy:
You give details about yourself — each key is the property, each value is its data.

```python
def show_profile(**details):
    for key, value in details.items():
        print(f"{key} : {value}")

show_profile(name="Memuna", role="Python Instructor", experience=2)
```

3️⃣ Lambda Expressions

- `lambda` function does not have any name.

- Can have any number of arguments but only one expression.

- lambda function returns a function

- Useful for short, simple functions.

- No need to write return statement.

- It can only contain expressions.

- you can use all the types of actual arguments.

```python
square = lambda x: x * x
print(square(5))  # 25
```

4️⃣ Nested Lambda

- Lambda inside another lambda.

- Can be used to return another function dynamically.

```python
nested_lambda = lambda x=5: (lambda y: x + y)
add = nested_lambda()
print(add(10))  # 15
```

5️⃣ Passing Lambda as Argument

- Functions can accept other functions as parameters.

- Lambda is often passed for inline behavior.

```python
def show(func):
    return func(8)

result = show(lambda x: x * 2)
print(result)  
```

6️⃣ Returning Lambda from Function

- Functions can return a lambda for delayed computation or closures.

```python
def multiplier(n):
    return lambda x: x * n

double = multiplier(2)
print(double(10))  # 20

triple = multiplier(3)
print(triple(10))  # 30
```

7️⃣ IIFE (Immediately Invoked Function Expression)

- Lambda or function called immediately after definition.

Analogy:

You order something online and immediately use it without storing it for later.

```python
result = (lambda x, y: x + y)(10, 20)
print(result)  # 30
```

📝 Practice Questions

1. Write a function that accepts any number of arguments and returns their product using `*args`.

2. Write a function using `**kwargs` to print employee details.

3. Create a lambda to find the cube of a number.

4. Write a nested lambda to add two numbers and then multiply by another number.

5. Pass a lambda to a function that doubles a given number.

6. Return a lambda from a function that multiplies input by 5.

7. Write an IIFE that subtracts two numbers and prints the result.

```python
# =========================
# Solution 1
# =========================
# Variable Length Arguments (*args)
def product(*nums):
    result = 1
    for n in nums:
        result *= n
    return result

print("Solution 1:", product(2, 3, 4))  # 24
print("Solution 1:", product(5, 10))    # 50


# =========================
# Solution 2
# =========================
# Keyword Variable Length Arguments (**kwargs)
def employee_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")

print("Solution 2:")
employee_details(name="Ali", role="Developer", experience=3)


# =========================
# Solution 3
# =========================
# Lambda Expression
cube = lambda x: x**3
print("Solution 3:", cube(3))  # 27


# =========================
# Solution 4
# =========================
# Nested Lambda
nested = lambda x: (lambda y: (x + y) * 2)
func = nested(5)
print("Solution 4:", func(10))  # (5+10)*2 = 30


# =========================
# Solution 5
# =========================
# Passing Lambda as Argument
def apply(func, value):
    return func(value)

print("Solution 5:", apply(lambda x: x*2, 7))  # 14


# =========================
# Solution 6
# =========================
# Returning Lambda
def multiplier(n):
    return lambda x: x*n

double = multiplier(2)
triple = multiplier(3)

print("Solution 6:", double(10))  # 20
print("Solution 6:", triple(10))  # 30


# =========================
# Solution 7
# =========================
# IIFE (Immediately Invoked Function Expression)
result = (lambda x, y: x - y)(15, 5)
print("Solution 7:", result)  # 10
```
