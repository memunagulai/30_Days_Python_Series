## Bitwise Operators
Operate on integer bits:

- `&` , `|` , `^` , `~` , `<<` , `>>`




```python
a = 5   # binary 0101
b = 3   # binary 0011

print("a & b ->", a & b)   # 1 (0001)
print("a | b ->", a | b)   # 7 (0111)
print("a ^ b ->", a ^ b)   # 6 (0110) XOR
print("~a ->", ~a)         # bitwise NOT (two's complement)
print("a << 1 ->", a << 1) # left shift (01010 => 10)
print("b >> 1 ->", b >> 1) # right shift (0001 => 1)
```

## 5) Membership Operators
Check presence in a sequence:

- `in` : True if element exists in container
- `not in` : True if element does not exist

**Examples**
- `"apple" in fruits_list`
- `value not in dict_keys`


```python
fruits = ["apple", "banana", "mango"]
print("apple in fruits ->", "apple" in fruits)
print("orange not in fruits ->", "orange" not in fruits)

text = "hello world"
print("'wor' in text ->", "wor" in text)
```

## 6) Identity Operators
Compare objects by **memory location**, not value:

- `is` : True if both are the same object
- `is not` : True if they are different objects

**Important note:** `is` is not for value equality ‚Äî use `==` for values.


```python
x = [1, 2, 3]
y = x        # y references same object
z = [1, 2, 3]# different object with same contents

print("x is y ->", x is y)   # True
print("x is z ->", x is z)   # False
print("x == z ->", x == z)   # True (values equal)
```

## 7) Operator Precedence & Associativity
Determines order of evaluation when many operators appear in an expression.

**Some key precedence (high ‚Üí low)**  
1. Parentheses `()`  
2. Exponent `**` (right-to-left associativity)  
3. `*`, `/`, `//`, `%` (left-to-right)  
4. `+`, `-` (left-to-right)  
5. Comparison operators (`<`, `>`, `==`, ...)  
6. `not`  
7. `and`  
8. `or`

**Tip:** Use parentheses `()` to make expressions clear and avoid mistakes.


```python
print("2 + 3 * 4 =", 2 + 3 * 4)       # 14
print("(2 + 3) * 4 =", (2 + 3) * 4)   # 20

# Right-to-left for **
print("2 ** 3 ** 2 =", 2 ** 3 ** 2)   # 512 because 3**2=9 -> 2**9

# Associativity left-to-right
print("10 - 4 - 1 =", 10 - 4 - 1)     # 5  -> (10-4)-1
```
