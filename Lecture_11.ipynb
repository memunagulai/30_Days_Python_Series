## 1) If-Else Statement
- `if-else` allows **two paths**: one executed if condition is True, another if condition is False.

**Syntax:**
```python
if condition:
    # code if condition True
else:
    # code if condition False


Real-Life Examples:

If you have an umbrella â†’ go outside, else â†’ stay inside

If exam marks >= 50 â†’ pass, else â†’ fail

````python
# Simple If-else
marks = 40
passing_marks = 50

if marks >= passing_marks:
    print("Congratulations! You passed.")
else:
    print("Sorry, you failed the exam.")
````

## 2) Nested If-Else
- You can **place an if-else inside another if-else**.  
- Useful for **multi-level decision making**.

**Example:**  
- If student passed, check marks to give grades:
    - Marks >= 90 â†’ Excellent  
    - Marks >= 75 â†’ Very Good  
    - Else â†’ Good


````python
marks = 85
passing_marks = 50

if marks >= passing_marks:
    print("You passed!")
    if marks >= 90:
        print("Excellent!")
    elif marks >= 75:
        print("Very Good!")
    else:
        print("Good!")
else:
    print("You failed the exam.")
````

## 3) If-Elif
- `elif` stands for "else if"  
- Used when we have **multiple conditions** to check sequentially.

**Syntax:**
```python
if condition1:
    # code1
elif condition2:
    # code2
elif condition3:
    # code3


Real-Life Example:

Temperature check:

30 â†’ Hot

20 â†’ Warm

10 â†’ Cool

````python
# If-Elif Example

temperature = 25

if temperature > 30:
    print("Hot day")
elif temperature > 20:
    print("Warm day")
elif temperature > 10:
    print("Cool day")
````

## 4) If-Elif-Else
- Combines **if**, **multiple elif**, and **else** to cover **all possibilities**.

**Real-Life Example:**  
- Grading system:
    - Marks >= 90 â†’ A+  
    - Marks >= 75 â†’ A  
    - Marks >= 50 â†’ B  
    - Else â†’ Fail


````python
marks = 78

if marks >= 90:
    print("Grade: A+")
elif marks >= 75:
    print("Grade: A")
elif marks >= 50:
    print("Grade: B")
else:
    print("Grade: Fail")
````

## Example: Age & Membership Access
- Age < 18 â†’ Access Denied  
- Age >= 18 & Membership â†’ Access Granted  
- Age >= 18 & No Membership â†’ Membership Required


````python
age = 20
has_membership = False

if age < 18:
    print("Access Denied")
elif age >= 18 and has_membership:
    print("Access Granted")
else:
    print("Membership Required")
````
