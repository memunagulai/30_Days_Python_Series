### **Pattern 15**

```python
rows = int(input("Enter number of rows: "))
i = 1
while i <= rows:
    j = 1
    start = chr(64 + rows - i + 1)   # starting character for row
    while j <= i:
        print(chr(ord(start) + j - 1), end=" ")
        j = j + 1
    print()
    i = i + 1
```
