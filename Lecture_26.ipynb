# Lecture 26 – File Handling

1️⃣ What is File Handling

File Handling allows a program to:

- Create files

- Read data from files

- Write data into files

- Update existing files

Why we need it?

Because data should persist even after the program ends.

Real-life example:
Saving notes in a notebook 📓

- RAM → your brain (temporary)

- File → notebook (permanent)

2️⃣ Types of Files

Python mainly works with two types of files:

🟢 Text Files

Human-readable

Examples: `.txt`, `.csv`, `.json`

🔵 Binary Files

Not human-readable

Examples: images, videos, .pdf, .exe

3️⃣ Opening a File

To work with a file, we must open it first.

file = open("filename", "mode")

```python
file = open("example.txt", "r")
print(file)
```

4️⃣ File Modes

| Mode | Meaning                      |
| ---- | ---------------------------- |
| `r`  | Read (file must exist)       |
| `w`  | Write (creates / overwrites) |
| `a`  | Append (adds data)           |
| `r+` | Read + Write                 |


5️⃣ Reading File

```python
file = open("example.txt", "r")
content = file.read()
print(content)
file.close()
```

🔹 read()

Reads entire file at once

```python
file = open("example.txt", "r")
print(file.read())
file.close()
```

🔹 readline()

Reads one line at a time

```python
file = open("example.txt", "r")
print(file.readline())
print(file.readline())
file.close()
```

🔹 readlines()

Reads all lines into a list

```python
file = open("example.txt", "r")
lines = file.readlines()
print(lines)
file.close()
```

6️⃣ Writing File

`w` mode creates or overwrites a file.

```python
file = open("data.txt", "w")
file.write("Hello Python!\n")
file.write("File Handling is easy.")
file.close()
```

7️⃣ Appending File

`a` mode adds data at the end of file.

```python
file = open("data.txt", "a")
file.write("\nThis line is appended.")
file.close()
```

8️⃣ with Statement

The best and safest way to work with files.

✔ Automatically closes the file

✔ Cleaner code

✔ No memory leaks

```python
with open("data.txt", "r") as file:
    print(file.read())
```

Real-life analogy:

Opening a door 🚪 and it automatically closes after you leave.

9️⃣ Closing File

If not using with, you must close the file manually.

file.close()

⚠️ Not closing files can cause:

- Memory issues

- File corruption

📝 Practice Questions

1. Create a text file and write your name and age into it

2. Read the file using `read()`

3. Read the same file line by line using `readline()`

4. Append a new line with your favorite programming language

5. Use `with` statement to read the final content

```python
# =========================
# Solution 1
# =========================
# Create a file and write name & age
file = open("student.txt", "w")
file.write("Name: Memuna Gul\n")
file.write("Age: 22\n")
file.close()


# =========================
# Solution 2
# =========================
# Read file using read()
file = open("student.txt", "r")
print("Solution 2:")
print(file.read())
file.close()


# =========================
# Solution 3
# =========================
# Read file line by line using readline()
file = open("student.txt", "r")
print("Solution 3:")
print(file.readline())
print(file.readline())
file.close()


# =========================
# Solution 4
# =========================
# Append favorite programming language
file = open("student.txt", "a")
file.write("Favorite Language: Python\n")
file.close()


# =========================
# Solution 5
# =========================
# Read final content using with statement
print("Solution 5:")
with open("student.txt", "r") as file:
    print(file.read())
```
