1️⃣ Functional Programming Concept

Functional Programming is a programming style where:

- Functions are treated as first-class citizens

- We focus on what to do, not how to do

- Functions avoid changing data (immutability)

Python supports functional programming using:

`map()`

`filter()`

`reduce()`

`lambda functions`

Real-life idea:

Think of an assembly line:

One worker transforms items → map

One worker removes defective items → filter

One worker combines everything into one result → reduce

2️⃣ `map()` Function

`map()` applies a function to each item in an iterable.

map(function, iterable)

It returns a map object, so we usually convert it to a list.

```python
# map() basic example
numbers = [1, 2, 3, 4, 5]

squared_numbers = list(map(lambda x: x ** 2, numbers))
print(squared_numbers)
```

Real-Life Example (map)

You have a list of prices, and you want to add 10% tax to each.

```python
prices = [100, 200, 300]

final_prices = list(map(lambda p: p + p * 0.10, prices))
print(final_prices)
```

3️⃣ `filter()` Function

`filter()` selects elements from an iterable based on a condition.

filter(function, iterable)

The function must return True or False.

```python
# filter() basic example
numbers = [1, 2, 3, 4, 5, 6]

even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)
```

Real-Life Example (filter)

From a list of students’ marks, select only those who passed (>= 50).

```python
marks = [45, 67, 89, 34, 55, 90]

passed_students = list(filter(lambda m: m >= 50, marks))
print(passed_students)
```

4️⃣ `reduce()` Function

`reduce()` reduces an iterable to a single value by repeatedly applying a function.

⚠️ `reduce()` is not built-in — it must be imported.

5️⃣ Importing `reduce` from `functools`

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

sum_of_numbers = reduce(lambda a, b: a + b, numbers)
print(sum_of_numbers)
```

Real-Life Example (reduce)

Calculate total bill amount from individual item prices.

```python
prices = [250, 300, 150, 400]

total_bill = reduce(lambda a, b: a + b, prices)
print(total_bill)
```

6️⃣ Difference Between map(), filter(), reduce()

| Function   | Purpose                            | Output            |
| ---------- | ---------------------------------- | ----------------- |
| `map()`    | Transform each element             | New iterable      |
| `filter()` | Select elements based on condition | Filtered iterable |
| `reduce()` | Combine all elements               | Single value      |


One-line summary:

- map → changes

- filter → selects

- reduce → combines

7️⃣ Combined Real-World Example

Scenario:

From a list of salaries:

Select salaries above 50,000

Increase them by 10%

Calculate total salary payout

```python
from functools import reduce

salaries = [30000, 50000, 70000, 90000, 40000]

filtered = filter(lambda s: s > 50000, salaries)
mapped = map(lambda s: s + s * 0.10, filtered)
total_payout = reduce(lambda a, b: a + b, mapped)

print(total_payout)
```

📝 Practice Questions

1. Use `map()` to convert a list of temperatures from Celsius to Fahrenheit

2. Use `filter()` to extract prime numbers from a list

3. Use `reduce()` to find the maximum number in a list

4. Combine `map()` and `filter()` to square only even numbers

5. Explain a real-world use case of `reduce()` in your own words

```python
from functools import reduce

# =========================
# Solution 1
# =========================
# Celsius to Fahrenheit using map()
celsius = [0, 20, 30, 40]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print("Solution 1:", fahrenheit)


# =========================
# Solution 2
# =========================
# Filter prime numbers using filter()
numbers = [2, 3, 4, 5, 6, 7, 8, 9, 11]

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

primes = list(filter(is_prime, numbers))
print("Solution 2:", primes)


# =========================
# Solution 3
# =========================
# Find maximum number using reduce()
nums = [10, 45, 23, 89, 67]
maximum = reduce(lambda a, b: a if a > b else b, nums)
print("Solution 3:", maximum)


# =========================
# Solution 4
# =========================
# Square only even numbers using filter + map()
nums = [1, 2, 3, 4, 5, 6]
even_squares = list(map(lambda x: x ** 2, filter(lambda x: x % 2 == 0, nums)))
print("Solution 4:", even_squares)


# =========================
# Solution 5
# =========================
# Real-world reduce example: total bill
prices = [1200, 850, 430, 999]
total_bill = reduce(lambda a, b: a + b, prices)
print("Solution 5:", total_bill)
```
