## üîπ Recap: Variables
A **variable** is a container that stores data. You can change the value anytime.

**Example:**
```python
age = 20
name = "Ali"
```
Daily Life Analogy: A jar can hold candies, and you can replace them anytime.

## 1Ô∏è‚É£ What are Data Types?
A **data type** tells Python what kind of value a variable holds.

**Daily Life Analogy:**
- Jar of candies = integer
- Jar of water = float
- Jar of letters = string
- Jar of yes/no switches = boolean

### Numbers
Python has different types of numbers:
- `int` ‚Üí Whole numbers, e.g., 5, 100
- `float` ‚Üí Decimal numbers, e.g., 3.14, 9.5
- `complex` ‚Üí Complex numbers, e.g., 2 + 3j (optional for beginners)

````
# Examples of Numbers
age = 25       # int
height = 5.9   # float
z = 2 + 3j     # complex

print("Age:", age)
print("Height:", height)
print("Complex Number:", z)
````

### Strings
**String (`str`)** is text inside quotes.

Example:
```python
name = "Sara"
city = 'Karachi'
```
Daily Life Analogy: Your name or city name written on paper.

````
# String Examples
name = "Sara"
city = 'Karachi'
print("Name:", name)
print("City:", city)
````

### Boolean
**Boolean (`bool`)** has two values: `True` or `False`

Example:
```python
is_student = True
is_teacher = False
```
Daily Life Analogy: Light switch ON/OFF, Yes/No answers.

````
# Boolean Example
is_student = True
is_teacher = False
print("Is student:", is_student)
print("Is teacher:", is_teacher)
````

### List
A **list** can store multiple values.

Example:
```python
grocery = ["milk", "eggs", "bread"]
```
Daily Life Analogy: Your shopping list.

````
# List Example
grocery = ["milk", "eggs", "bread"]
print("Grocery List:", grocery)
````

## 2Ô∏è‚É£ Declaring & Using Variables
You don't need to declare types in Python, it's dynamic.

**Example:**
```python
x = 10        # int
x = "Hello"  # now string
```
Daily Life Analogy: The same jar can hold candies today, cookies tomorrow.

````
# Variable usage example
x = 10
print("x:", x, type(x))

x = "Hello"
print("x now:", x, type(x))
````

## 3Ô∏è‚É£ Type Checking
Use `type()` to check the data type of a variable.

Example:
```python
num = 5
print(type(num))  # <class 'int'>
```

````
# Type checking example
num = 5
pi = 3.14
name = "Ali"
flag = True

print("Type of num:", type(num))
print("Type of pi:", type(pi))
print("Type of name:", type(name))
print("Type of flag:", type(flag))
````

## 4Ô∏è‚É£ Type Conversion (Casting)
Convert one type to another using:
- `int()`
- `float()`
- `str()`

Example:
```python
num_str = "10"
num_int = int(num_str)
```

````
# Type conversion examples
num_str = "10"
num_int = int(num_str)
print("String:", num_str, type(num_str))
print("Integer:", num_int, type(num_int))

pi = 3.14
pi_int = int(pi)
print("Float to int:", pi_int, type(pi_int))
````

## 5Ô∏è‚É£ Common Beginner Mistakes
- Mixing int and str:
```python
age = 20
print("Age: " + age)  # ‚ùå Error
```
Correct:
```python
print("Age: " + str(age))  # ‚úÖ
```
- Case sensitivity: `Age` ‚â† `age`
