# Lecture 14 ‚Äì Pattern Problems in Python

This lecture focuses on logic building using loops.

Patterns are not about stars ‚≠ê they are about understanding nested loops and how rows & columns work.

**Core Concept (Very Important)**

Outer loop ‚Üí Rows

Inner loop ‚Üí Columns

Pattern size is controlled using a variable like n



üîπ Pattern 1: Square Pattern (Nested for Loop)

```python
n = 5
for i in range(n): # 0,1,2,3,4
    for j in range(n):
        print("*", end=" ")
    print()
```


üîπ Pattern 2: Right Triangle (Nested for Loop)



```python
n = 5
for i in range(1, n + 1):
    for j in range(i):          
        print("*", end=" ")
    print()
```

üîπ Pattern 3: Inverted Right Triangle (Nested for Loop)

```python
n = 5
for i in range(n, 0, -1):
    for j in range(i):
        print("*", end=" ")
    print()
```

üîπ Pattern 4: Number Triangle

```python
n = 5
for i in range(1, n + 1): 
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

üîπ Pattern 5: Repeating Number Triangle

```python
n = 5
for i in range(1, n + 1):
    for j in range(i):
        print(i, end=" ")
    print()
```


üîπ Pattern 6: Square Pattern (Nested while Loop)

```python
n = 5
i = 0
while i < n:
    j = 0
    while j < n:
        print("*", end=" ")
        j += 1
    print()
    i += 1
```

üîπ Pattern 7: Right Triangle (Nested while Loop)

```python
n = 5
i = 1
while i <= n:
    j = 1
    while j <= i:
        print("*", end=" ")
        j += 1
    print()
    i += 1
```

üîπ Pattern 8: Inverted Triangle (Nested while Loop)

```python
n = 5
i = n
while i > 0:
    j = 0
    while j < i:
        print("*", end=" ")
        j += 1
    print()
    i -= 1
```

üîπ Pattern 9: Alphabet Pattern

```python
n = 5
for i in range(n):
    for j in range(i + 1):
        print(chr(65 + i), end=" ")
    print()
```

üîπ Pattern 10: Continuous Number Pattern

```python
num = 1
n = 4
for i in range(1, n + 1):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```
