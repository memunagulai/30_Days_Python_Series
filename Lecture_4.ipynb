## 1Ô∏è‚É£ Identifiers
An **identifier** is a name you give to a variable, function, class, or module in Python.

**Rules:**
- Can include letters, digits, and underscore (_)
- Must start with a letter or underscore
- Case-sensitive
- Cannot be a reserved word
- Should be meaningful

**Daily Life Example:**
- Your name, your age, and your school name are identifiers.

````
# Examples of valid identifiers
name = "Ali"
_age = 21
school_name = "Sunrise School"

print(name)
print(_age)
print(school_name)
````

**Invalid Identifiers Examples**
- `2name` ‚ùå starts with a number
- `user-name` ‚ùå contains hyphen
- `class` ‚ùå reserved word

**Tip:** Use meaningful names like `student_name` instead of `x1`.

## 2Ô∏è‚É£ Reserved Words (Keywords)
Reserved words are **special words** that Python already uses. You **cannot use them** as identifiers.

**Examples:**
```
if, else, for, while, class, import, True, False, def
```
**Daily Life Analogy:** Reserved words are like street signs ‚Äî you cannot make your own ‚ÄúStop‚Äù sign differently.

````
import keyword
print("Python Reserved Words / Keywords:")
print(keyword.kwlist)
````

## 3Ô∏è‚É£ Variables
A **variable** is a container to store data.
‚úÖ Can change value anytime
‚úÖ No need to declare type in Python (Dynamic typing)

**Daily Life Analogy:** A variable is like a jar. You can put candies (value) in it, and later replace them.

````
# Example of variables
age = 20
print("Age:", age)

age = 25  # Value updated
print("Updated Age:", age)

name = "Sara"
print("Name:", name)
````

## 4Ô∏è‚É£ Constants
A **constant** is a value that **should not change**. Python does not enforce constants strictly, but by convention we use **ALL_CAPS**.

**Examples:**
```
PI = 3.14159
MAX_STUDENTS = 40
```
**Daily Life Analogy:** Birthdate, school founding year ‚Äî values that should remain constant.

````
# Example of constants (by convention)
PI = 3.14159
MAX_STUDENTS = 40

print("PI =", PI)
print("Maximum Students =", MAX_STUDENTS)
````

# ‚úÖ Quick Recap
- **Identifier** = Name for variables, functions, classes
- **Reserved Word / Keyword** = Cannot use as identifier
- **Variable** = Container for data (can change)
- **Constant** = Fixed value (should not change)

Daily Life Analogy:
- Jar = Variable
- Birthdate = Constant
- Street Signs = Reserved Words
- Your Name = Identifier
