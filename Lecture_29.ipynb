# Lecture 29 – OOPs in Python (Methods & Encapsulation)

1️⃣ Instance Methods

- Most common type of method

- Works on object data

- Uses `self` as first parameter

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):   # instance method
        print(self.name, self.marks)

s1 = Student("Ali", 85)
s1.display()
```

2️⃣ Accessor (Getter) Method

- Used to read data

- Provides controlled access

```python
class Student:
    def __init__(self, marks):
        self.marks = marks   # instance variable

    def get_marks(self):
        return self.marks

s = Student(90)
print(s.get_marks())
```

3️⃣ Mutator (Setter) Method

- Used to modify data

- Allows validation before update

```python
class Student:
    def __init__(self, marks):
        self.marks = marks

    def set_marks(self, marks):
        if marks >= 0:
            self.marksarks = marks

    def get_marks(self):
        return self.marks

s = Student(80)
s.set_marks(95)
print(s.get_marks())
```

4️⃣ Encapsulation in Python

Encapsulation means:

- Binding data and methods together

- Restricting direct access to data

Achieved using:

- Private variables (__var)

- Getter & Setter methods

📌 Think of it as a capsule 💊 protecting data.

5️⃣ Private & Protected Members (Conceptual)

| Type      | Syntax   | Meaning                     |
| --------- | -------- | --------------------------- |
| Public    | `name`   | Accessible everywhere       |
| Protected | `_name`  | Access within class & child |
| Private   | `__name` | Access only inside class    |


```python
class Demo:
    
    def __init__(self):
        self.public = 10
        self._protected = 20
        self.__private = 30

    def show(self):
        print("Inside class:")
        print(self.public)
        print(self._protected)
        print(self.__private)


d = Demo()

# Accessing outside the class
print("Public:", d.public)        # Accessible

print("Protected:", d._protected) # Not recommended but accessible

print("Private:", d.__private)    # Will give error
```

6️⃣ Class Method

- Works with class variables

- Uses `@classmethod`

- First parameter is `cls`

```python
class Student:
    institute = "ABC Institute"

    @classmethod
    def change_institute(cls, name):
        cls.institute = name

Student.change_institute("XYZ Institute")  #calling class method
print(Student.institute)
```

7️⃣ Static Method

- No `self`, no `cls`

- Utility/helper method

- Uses `@staticmethod`

```python
class MathUtils:
    @staticmethod
    def add(a, b):
        return a + b

print(MathUtils.add(5, 3))   #calling static method
```

- Static Methods are used when some processing is related to the class but does not need any class or instances to perform any work

8️⃣ Difference: Instance vs Class vs Static Method

| Feature   | Instance        | Class          | Static          |
| --------- | --------------- | -------------- | --------------- |
| Parameter | `self`          | `cls`          | None            |
| Access    | Object data     | Class data     | No data         |
| Decorator | None            | `@classmethod` | `@staticmethod` |
| Usage     | Object behavior | Class behavior | Utility logic   |


9️⃣ Passing Member of One Class to Another Class

Objects can be passed as arguments.

```python
class Student:
    #constructor
    def __init__(self, name, rollno):
        self.name = name
        self.roll= rollno
    
    # Instance Method
    def disp(self):
        print("Student Name",self.name)
        print("Student Roll", self.roll)

class User:
    # Static Method
    @staticmethod
    def show(s):
        print("Username",s.name)
        print("RollNo",s.roll)
        s.disp()

stu = Student("Ali",14)
User.show(stu)
```

🔟 Nested Class

A class inside another class.

Used when:

One class is completely dependent on another

```python
class University:

    def __init__(self, uni_name, location, dept_name, head):
        self.uni_name = uni_name
        self.location = location
        
        # Creating object of inner class
        self.dept = self.Department(dept_name, head)

    def show_university(self):
        print("University:", self.uni_name)
        print("Location:", self.location)

    class Department:

        def __init__(self, dept_name, head):
            self.dept_name = dept_name
            self.head = head

        def show_department(self):
            print("Department:", self.dept_name)
            print("Head:", self.head)


# Creating object
u = University("ABC University", "Islamabad", "Computer Science", "Dr Ali")

u.show_university()
u.dept.show_department()
```

📝 Practice Questions

1. Create a class with private variable and getter/setter

2. Create a class method to update a class variable

3. Write a static method for validation logic

4. Pass object of one class into another class

5. Create and access a nested class

```python
# =========================
# Solution 1
# =========================
# Private variable with getter and setter
class Student:
    def __init__(self, marks):
        self.__marks = marks

    def get_marks(self):
        return self.__marks

    def set_marks(self, marks):
        if marks >= 0:
            self.__marks = marks

s = Student(80)
s.set_marks(95)
print("Solution 1:", s.get_marks())


# =========================
# Solution 2
# =========================
# Class method to update class variable
class Student:
    institute = "ABC Institute"

    @classmethod
    def change_institute(cls, name):
        cls.institute = name

Student.change_institute("XYZ Institute")
print("Solution 2:", Student.institute)


# =========================
# Solution 3
# =========================
# Static method for validation
class Validator:
    @staticmethod
    def is_positive(num):
        return num > 0

print("Solution 3:", Validator.is_positive(10))
print("Solution 3:", Validator.is_positive(-5))


# =========================
# Solution 4
# =========================
# Passing object of one class to another
class Student:
    def __init__(self, name):
        self.name = name

class Teacher:
    def show_student(self, student):
        print("Solution 4: Student name is", student.name)

s = Student("Ali")
t = Teacher()
t.show_student(s)


# =========================
# Solution 5
# =========================
# Nested class example
class University:
    class Department:
        def show(self):
            print("Solution 5: Computer Science Department")

dept = University.Department()
dept.show()
```
