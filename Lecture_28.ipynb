1️⃣ What is Object Oriented Programming (OOP)

Object Oriented Programming (OOP) is a programming paradigm where we organize code using:

- Objects (real-world entities)

- Classes (blueprints)

Instead of writing logic only with functions, we model data + behavior together.

2️⃣ Why OOP is Needed

OOP helps us to:

- Manage large codebases

- Improve code reusability

- Make programs easy to understand

- Reduce code duplication

- Model real-world problems naturally

Real-life analogy:

Without OOP → everything scattered

With OOP → everything organized into proper objects 🗂️

3️⃣ Real-World Example of OOP

- Student → Object

- Name, age, roll number → Data

- Study(), take_exam() → Behavior

Each student is an object created from a class.

4️⃣ Four Pillars of OOP

🔹 Encapsulation

Wrapping data + methods inside a class.

🔹 Abstraction

Showing only essential details, hiding complexity.

🔹 Inheritance

One class inherits properties of another.

🔹 Polymorphism

Same method name, different behavior.

7️⃣ Constructor (`__init__`)

- A special method

- Automatically called when object is created

- Used to initialize data

```python
#without parameter
class Student:
    
    def __init__(self):
        print("Student object created")

# Creating object
s1 = Student()
```

```python
#with parameter
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

student1 = Student("Ali", 20)
student2 = Student("Sara", 22)

print(student1.name, student1.age)
print(student2.name, student2.age)
```

8️⃣ Instance Variables

Variables belonging to an object

Created using `self`

Each object has its own copy

```python
class Student:
    def __init__(self, name):
        self.name = name   # instance variable
    def show_name(self):
        self.name  #access instance variable
s1 = Student("Ali")
s2 = Student("Sara")

print(s1.name)  # access instance variable outside class 
print(s2.name)
```

9️⃣ Class Variables (Static Variables)

- Variables shared by all objects

- Defined inside class, outside methods

```python
class Student:
    institute = "ABC Institute"   # class variable

    def __init__(self, name):
        self.name = name

s1 = Student("Ali")
s2 = Student("Sara")

print(s1.institute)
print(s2.institute)
```

🔟 Namespace Concept

A namespace is a space where names are stored.

Types:

- Instance namespace

- Class namespace



```python
class Demo:
    value = 100   # class namespace

    def __init__(self):
        self.value = 50   # instance namespace

obj = Demo()
print(obj.value)
```

- Class Namespace → A storage area that keeps all names defined in a class.

- Class Variable → A specific variable stored inside the class namespace and shared by all objects.

- Instance Namespace: A container that stores all attributes (variables) belonging to a specific object.

- Instance Variable: A variable inside the instance namespace that stores data unique to that object.

1️⃣1️⃣ Difference Between Instance & Class Variables

| Instance Variable             | Class Variable            |
| ----------------------------- | ------------------------- |
| Belongs to object             | Belongs to class          |
| Uses `self`                   | Defined directly in class |
| Separate copy for each object | Shared among all objects  |
| Object-specific data          | Common data               |

```python
# =========================
# Solution 1
# =========================
# Create a Car class with brand and year
class Car:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year


# =========================
# Solution 2
# =========================
# Create two objects and print details
car1 = Car("Toyota", 2020)
car2 = Car("Honda", 2022)

print("Solution 2:")
print(car1.brand, car1.year)
print(car2.brand, car2.year)


# =========================
# Solution 3
# =========================
# Add class variable wheels
class Car:
    wheels = 4  # class variable

    def __init__(self, brand, year):
        self.brand = brand
        self.year = year

car1 = Car("Toyota", 2020)
car2 = Car("Honda", 2022)

print("Solution 3:")
print(car1.wheels)
print(car2.wheels)


# =========================
# Solution 4
# =========================
# Difference between instance and class variables
class Student:
    institute = "ABC Institute"  # class variable

    def __init__(self, name):
        self.name = name  # instance variable

s1 = Student("Ali")
s2 = Student("Sara")

print("Solution 4:")
print(s1.name, s1.institute)
print(s2.name, s2.institute)


# =========================
# Solution 5
# =========================
# Real-world OOP example
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        self.balance -= amount

account = BankAccount("Memuna", 1000)
account.deposit(500)
account.withdraw(200)

print("Solution 5:")
print(account.owner, account.balance)
```
