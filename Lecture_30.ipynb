# Lecture 30 – OOP in Python (Inheritance & Polymorphism)

1️⃣ What is Inheritance

Inheritance allows one class (child) to acquire properties and methods of another class (parent).

📌 Promotes code reusability and logical hierarchy.

All classes in python are built from a single parent class called `object`

- Parent class is also known as super class , Base class

- child class is also known as sub class , derived class

2️⃣ Why We Need Inheritance

- Avoids code duplication

- Improves maintainability

- Represents real-world relationships

(Car → ElectricCar, Employee → Manager)

3️⃣ Types of Inheritance

🔹 Single Inheritance

- If a class is derived from one base class(parent class) then its calles single inheritance.

- we can access parent class variables and methods using Child class object

- we can also access parent class variables and methods using parent class object

- we can not access child class variables and methods using parent class object

One parent → one child

```python
class Animal:
    def eat(self):
        print("Animal eats")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

d = Dog()
d.eat()
d.bark()
```

```python
# Single Inheritance

class Father:
    money = 1000

    def show(self):
        print("Parent Class Instance Method")

    @classmethod
    def showmoney(cls):
        print("Parent Class Class Method:", cls.money)


class Son(Father):
    def disp(self):
        print("Child Class Instance Method")


s = Son()
s.disp()
s.show()
s.showmoney()
```

🔹 Multilevel Inheritance

Parent → Child → Grandchild

```python
class Vehicle:
    def start(self):
        print("Vehicle starts")

class Car(Vehicle):
    def drive(self):
        print("Car drives")

class ElectricCar(Car):
    def charge(self):
        print("Electric car charging")

e = ElectricCar()
e.start()
e.drive()
e.charge()
```

🔹 Hierarchical Inheritance

One parent → multiple children

```python
# Hierarchical Inheritance

class Father:
    money = 1000

    def show(self):
        print("Parent Class Method")


class Son(Father):
    def disp_son(self):
        print("Son Class Method")


class Daughter(Father):
    def disp_daughter(self):
        print("Daughter Class Method")


s = Son()
s.disp_son()
s.show()

print()

d = Daughter()
d.disp_daughter()
d.show()
```

```python
class Shape:
    def draw(self):
        print("Drawing shape")

class Circle(Shape):
    pass

class Square(Shape):
    pass

Circle().draw()
Square().draw()
```

🔹 Multiple Inheritance

Child inherits from multiple parents

```python
# Parent classes
class Father:
    def f_skills(self):
        print("Father's skills: Gardening")

class Mother:
    def m_skills(self):
        print("Mother's skills: Cooking")

# Child class inherits from both Father and Mother
class Child(Father, Mother):
    def c_skills(self):
        print("Child's skills:")
        print(" Programming skills!")

# Create instance of Child
c = Child()
c.c_skills()
c.f_skills()
c.m_skills()
```

```python
# Parent classes
class Father:
    def __init__(self):
        super().__init__()
        print("Father Constructor")
    def f_skills(self):
        print("Father's skills: Gardening")

class Mother:
    def __init__(self):
        super().__init__()
        print("Mother Constructor")
    def m_skills(self):
        print("Mother's skills: Cooking")

# Child class inherits from both Father and Mother
class Child(Father, Mother):
    def __init__(self):
        super().__init__()
        print("Child Constructor")
    def c_skills(self):
        print("Child's skills:")
        print(" Programming skills!")

# Create instance of Child
c = Child()
```

4️⃣ Constructor in Inheritance

- The constructor in parent class is available to the child class

```python
class Parent:
    def __init__(self):
        self.money=5000
        print("Parent constructor")

class Child(Parent):
    def disp(self):
        print("Child method")

ch= Child()
ch.money
```

5️⃣ Constructor Overriding & super()

- If we write constructors in both classes , parent class and child class then the parent class constructor is not available to the child class

- Constructor overriding is used when programmer want to modify the existing behavior of a constructor

```python
class Parent:
    def __init__(self):
        print("Parent constructor")

class Child(Parent):
    def __init__(self):
        print("Child constructor")

ch= Child()
```

```python
class Parent:
    def __init__(self):
        print("Parent constructor")

class Child(Parent):
    def __init__(self):
        super().__init__()
        print("Child constructor")

ch = Child()
```

`super()` calls parent class constructor/method from a child class

6️⃣ Method Resolution Order (MRO)

Defines order in which methods are searched.

- In the Multiple Inheritance senario members of class are searched first in the current class.if not found, the search continues into parent classes in depth-first, left to right manner without searching the same class twice.

```python
# Parent classes
class Father:
    def __init__(self):
        super().__init__()
        print("Father Constructor")
    def f_skills(self):
        print("Father's skills: Gardening")

class Mother:
    def __init__(self):
        super().__init__()
        print("Mother Constructor")
    def m_skills(self):
        print("Mother's skills: Cooking")

# Child class inherits from both Father and Mother
class Son(Father, Mother):
    def __init__(self):
        super().__init__()
        print("Child Constructor")
    def c_skills(self):
        print("Child's skills:")
        print(" Programming skills!")

# Create instance of Child
c = Son()
```

7️⃣ What is Polymorphism

“One name, many forms”

📌 Same method behaves differently for different objects.

8️⃣ Method Overloading (Conceptual in Python)

Python does not support true overloading.

Achieved using:

- Default arguments

- Variable arguments

```python
class Math:
    def add(self, a, b=0, c=0):
        s= a + b + c
        return s

m = Math()
print(m.add(2))
print(m.add(5,6))
print(m.add(10,78,45))
```

9️⃣ Method Overriding

Child class redefines parent method.

```python
class Add:
    def result(self,a,b):
        print("Addition",a+b)

class Multi(Add):
    def result(self,a,b):
        print("Multiplication",a*b)

m= Multi()
m.result(12,6)
```

```python
class Add:
    def result(self,x,y):
        print("Addition",x+y)

class Multi(Add):
    def result(self,a,b):
        super().result(10,4)
        print("Multiplication",a*b)

m= Multi()
m.result(12,6)
```

🔟 Duck Typing

“If it walks like a duck and quacks like a duck…” it must b duck

Object type doesn’t matter — method behavior does.

```python
class Dog:
    def speak(self):
        print("Bark")

class Cat:
    def speak(self):
        print("Meow")

def animal_sound(animal):
    animal.speak()

d= Dog()
animal_sound(d)
c=Cat()
animal_sound(c)
```

1️⃣1️⃣ Strong Typing

Python is strongly typed (no implicit type mixing).

```python
class Dog:
    def speak(self):
        print("Bark")

class Cat:
    def speak(self):
        print("Meow")

def animal_sound(animal):
    if hasattr(animal,'speak'):
       animal.speak()

d= Dog()
animal_sound(d)
c=Cat()
animal_sound(c)
```

1️⃣2️⃣ Operator Overloading

- If any Operator perform additional actions other than what it is meant for, its called operator overloading.

```python
10+20

int.__add__(10,20)

int.__add__(self,other) # Magic function

a+b

str.__add__('a','b')
```

```python
class A:
    def __init__(self, x):
        self.x = x

    def __add__(self, other):
        sum= self.x + other.x
        return sum

class B:
    def __init__(self,x):
        self.x= x
a = A(10)
b = B(20)
print(a+b)
```

1️⃣3️⃣ Abstraction

Hides implementation details and shows only essential features.

Achieved using abstract classes.

1️⃣4️⃣ Abstract Class & Abstract Method

we use abstract class when there are some common feature shared by all the objects as they are.

- PVM cannot create object of an abstract class.

- Its not necessary to declare all methods abstract in a abstract class.

- Abstract class can have abstract method and concrete method.

- If there's any abstract method in a class , that class must b abstract.

- The abstract methods of an abstract class must be defined in its child class.



```python
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):  #abstract class is redefined in child classes
        pass

    def info(self):
        print("This is a shape")

class Rectangle(Shape):
    def area(self):
        return 10 * 5

r = Rectangle()
print(r.area())
r.info()
```

1️⃣5️⃣ Concrete Method

A concrete method is a method whose action is defined in the abstract class itself.

A fully implemented method inside abstract class

(example: `info()` above)

1️⃣6️⃣ Interface in Python (Conceptual)

Python has no true interface.

Achieved using:

- Abstract class with only abstract methods

1️⃣7️⃣ Difference: Abstract Class vs Interface

| Feature              | Abstract Class      | Interface (Conceptual) |
| -------------------- | ------------------- | ---------------------- |
| Methods              | Abstract + Concrete | Only Abstract          |
| Variables            | Allowed             | Constants only         |
| Multiple Inheritance | Yes                 | Yes                    |
| Constructor          | Allowed             | Not used               |
