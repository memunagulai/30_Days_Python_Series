## 1) If Statement
An `if` statement allows the program to **execute certain code only if a condition is True**.  

**Basic Syntax:**
```python
if condition:
    # code block executed if condition is True


Real-Life Examples:

If it is raining, take an umbrella.

If your marks are greater than passing marks, you pass the exam.

````python
marks = 75
passing_marks = 50

if marks >= passing_marks:
    print("Congratulations! You passed the exam.")

# This code prints message only if the condition is True
````

## 2) If Statement with Logical Operators
You can combine multiple conditions using `and`, `or`, `not`.

**Example:**
- If age >= 18 AND has_ID -> eligible to vote
- If score >= 90 OR extra_credit >= 5 -> student gets award


````python
age = 20
has_id = True

if age >= 18 and has_id:
    print("You are eligible to vote.")

score = 92
extra_credit = 3

if score >= 90 or extra_credit >= 5:
    print("You receive a special award!")
````

## 3) Importance of Indentation
- Python uses **indentation** (spaces or tabs) to define **code blocks**.
- **All statements inside if must be indented equally**.
- Improper indentation causes `IndentationError`.

**Example:**
```python
if True:
    print("Hello")
    print("World")  # Both lines are part of if-block


````python


# Correct indentation
if True:
    print("This line is inside the if block")
    print("This line is also inside the if block")

# Uncommenting below will raise error
# if True:
# print("This will cause IndentationError")
````

## 4) Nested If
- You can place one `if` statement **inside another**.  
- Useful for checking multiple levels of conditions.

**Real-Life Examples:**
- If student passed exam, then check if marks > 90 â†’ award gold star
- If weather is sunny, then check if temperature > 35 â†’ stay indoors


````python
marks = 95
passing_marks = 50

if marks >= passing_marks:
    print("You passed the exam!")
    if marks > 90:
        print("Excellent performance! Gold star awarded!")
````

## Example: Age Verification & Membership
- If user is age >= 18:
    - If user has membership -> print "Access granted"
    - Else -> print "Membership required"


````python
age = 20
has_membership = True

if age >= 18:
    print("Age requirement passed.")
    if has_membership:
        print("Access granted!")
    else:
        print("Membership required.")
````

````python
# Login simulation
age = int(input("Enter your age: "))
membership = input("Are you a member? (yes/no): ")

# Only if statements
if age >= 18 and membership == "yes":
    print("Access Granted")

if age >= 18 and membership == "no":
    print("Membership Required")

if age < 18:
    print("Access Denied")
````
