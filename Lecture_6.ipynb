## 🟦 Section 1: Output in Python (`print`)

```python
# Simple print example
print("Hello, Python!")
```

```python
# Print multiple values
print("My name is", "Memuna Gul")
```

## 🟦 Section 2: Input in Python (`input`)

```python
# Taking user's name
name = input("Enter your name: ")
print("Hello,", name)
```

```python
# Taking user's age
age = input("Enter your age: ")
print("You are", age, "years old.")
```

## 🟦 Section 3: Type Casting (Converting Input)

```python
# Adding two numbers correctly
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
print("Sum =", num1 + num2)
```

```python
# Working with decimal numbers
price = float(input("Enter price: "))
print("Price is:", price)
```

## 🟦 Section 4: F-Strings (Formatted Output)

```python
# Using f-strings for clean output
student = "Ali"
score = 95
print(f"Student: {student}, Score: {score}")
```

## 🟦 Section 5: Common Beginner Mistakes

```python
# ❌ Mistake: Adding strings as numbers
a = input("Enter number: ")
b = input("Enter number: ")
print("Wrong result:", a + b)  # This joins the strings
```

```python
# ✅ Correct way
print("Correct result:", int(a) + int(b))
```

## 🟦 Section 6: Full Example Program

```python
name = input("Enter your name: ")
city = input("Enter your city: ")
age = int(input("Enter your age: "))

print()
print("--- Profile Summary ---")
print(f"Name: {name}")
print(f"City: {city}")
print(f"Age: {age}")
```
