1Ô∏è‚É£ What is a Function?

A function is a block of reusable code that performs a specific task.

Instead of writing the same logic again and again, we define it once and reuse it whenever needed.

üìå Why functions are important (especially for AI):

AI systems are built using modular functions (data loading, preprocessing, training, prediction).

Functions help make code clean, readable, and reusable.

Real-life example:

Think of a function like a coffee machine button ‚òï

You press it ‚Üí coffee is made.

You don‚Äôt care how it works inside ‚Äî you just reuse it.

```python
def greet():
    print("Hello, welcome to Python!")

greet()
```

2Ô∏è‚É£ Built-in Functions

Built-in functions are functions that Python already provides.
You don‚Äôt need to define them just use them.

üìå Examples of common built-in functions:

`print()`

`len()`

`type()`

`sum()`

`max()`

`min()`

`input()`

```python
numbers = [10, 20, 30]

print(len(numbers))
print(sum(numbers))
print(max(numbers))
```

3Ô∏è‚É£ User-Defined Functions

User-defined functions are functions created by the programmer to solve a specific problem.

They are defined using the def keyword.

```python
def add(a, b):
    return a + b

print(add(5, 3))
```

4Ô∏è‚É£ Returning Single Value

A function can return one value using `return`.

```python
def square(num):
    return num * num

print(square(4))
```

5Ô∏è‚É£ Returning Multiple Values

Python allows returning multiple values from a function.

```python
def calculate(a, b):
    return a + b, a - b, a * b

add, sub, mul = calculate(10, 5)
print(add, sub, mul)
```

6Ô∏è‚É£ Nested Function

A nested function is a function inside another function.

```python
def outer():
    
    def inner():
        print("Inner function")
    print("Outer function")

    inner()

outer()
```

7Ô∏è‚É£ Passing Function as Parameter

Functions can be passed as arguments to other functions

```python
def disp(sh):
    print("Disp Function" + sh())

def show():
    return "Show Function"

disp(show)
```

```python
def disp(sh):
    return "Disp Function" + sh()

def show():
    return "Show Function"

result= disp(show)
print(result)
```

```python
def add(a, b):
    return a + b

def operate(func, x, y):
    return func(x, y)

print(operate(add, 3, 4))
```

8Ô∏è‚É£ Function Returning Function

A function can return another function.

```python
def outer():
    def inner():
        return "Hello from inner function"
    print("hello from outer function")
    return inner

res = outer()
print(res())
```

9Ô∏è‚É£ Actual vs Formal Arguments

Formal Arguments ‚Üí Variables in function definition

Actual Arguments ‚Üí Values passed during function call

```python
def greet(name):   # formal argument
    print("Hello", name)

greet("Alice")    # actual argument
```

üîü Positional Arguments

Arguments passed in sequence.

```python
def student(name, age):
    print(name, age)

student("Ali", 20)
```

‚ö† Order matters.

1Ô∏è‚É£1Ô∏è‚É£ Keyword Arguments

Arguments passed using key = value.

```python
student(age=20, name="Ali")
```

‚úî Order doesn‚Äôt matter.

1Ô∏è‚É£2Ô∏è‚É£ Default Arguments

Default values are used if no argument is provided.

```python
def greet(name="User"):
    print("Hello", name)

greet()
greet("Memuna")
```
