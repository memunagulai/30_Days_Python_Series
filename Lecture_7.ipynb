## 1Ô∏è‚É£ What are Operators?
- Operators are symbols that perform operations on values or variables.
- Daily Life Analogy: `+` is like adding candies, `-` is taking away.

## 2Ô∏è‚É£ Arithmetic Operators in Python
| Operator | Meaning | Example |
|----------|---------|---------|
| +        | Addition | 5 + 3 = 8 |
| -        | Subtraction | 10 - 4 = 6 |
| *        | Multiplication | 3 * 4 = 12 |
| /        | Division | 10 / 2 = 5.0 |
| //       | Floor Division | 10 // 3 = 3 |
| %        | Modulus | 10 % 3 = 1 |
| **       | Exponentiation | 2 ** 3 = 8 |

````
# Arithmetic Operators Examples
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponentiation:", a ** b)
````

## 3Ô∏è‚É£ Operator Precedence
- Python follows BODMAS:
  1. Parentheses ()
  2. Exponent **
  3. Multiplication *, Division /, Floor Division //, Modulus %
  4. Addition +, Subtraction -

**Example:**
```python
result = 2 + 3 * 4  # 2 + (3*4) = 14
```

````
# Operator Precedence Example
result = 2 + 3 * 4
print("2 + 3 * 4 =", result)

result2 = (2 + 3) * 4
print("(2 + 3) * 4 =", result2)
````

## 4Ô∏è‚É£ Mini Projects / Practical Examples
- **Simple Calculator**
- **Age Calculator**
- **Total Price Calculator**
- **BMI Calculator**

We'll take input from the user and use arithmetic operators.

````
# Simple Calculator Example
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print(f"Addition: {num1 + num2}")
print(f"Subtraction: {num1 - num2}")
print(f"Multiplication: {num1 * num2}")
print(f"Division: {num1 / num2}")
````

````
# Age Calculator Example
current_year = int(input("Enter current year: "))
birth_year = int(input("Enter your birth year: "))
age = current_year - birth_year
print(f"You are {age} years old.")
````

````
# Total Price Calculator Example
price = float(input("Enter price of one item: "))
quantity = int(input("Enter quantity: "))
total = price * quantity
print(f"Total price is {total}")
````

````
# BMI Calculator Example
weight = float(input("Enter weight in kg: "))
height = float(input("Enter height in meters: "))
bmi = weight / (height ** 2)
print(f"Your BMI is {bmi:.2f}")
````
